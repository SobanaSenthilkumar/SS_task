This notebook covers NumPy package with examples and practice problems.

```python
import numpy as np
print('NumPy Version:', np.__version__)
```

## 1ï¸ Creating NumPy Arrays

```python
# 1D Array
arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)

# 2D Array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2)

# 3D Array
arr3 = np.array([[[1,2],[3,4]], [[5,6],[7,8]]])
print(arr3)
```

## 2ï¸ Array Properties

```python
arr = np.array([[10, 20, 30], [40, 50, 60]])
print('Shape:', arr.shape)
print('Dimensions:', arr.ndim)
print('Size:', arr.size)
print('Data Type:', arr.dtype)
```

## 3ï¸ Special Arrays

```python
print(np.zeros((2,3)))
print(np.ones((3,3)))
print(np.eye(3))
print(np.arange(0, 10, 2))
print(np.linspace(0, 1, 5))
print(np.random.rand(3,3))
print(np.random.randint(1, 100, (3,3)))
```

## 4ï¸ Indexing and Slicing

```python
arr = np.array([10, 20, 30, 40, 50])
print(arr[0])
print(arr[-1])
print(arr[1:4])

arr2 = np.array([[1,2,3],[4,5,6]])
print(arr2[0,1])
print(arr2[:,1])
```

## 5ï¸ Reshaping Arrays

```python
arr = np.arange(1, 13)
reshaped = arr.reshape(3,4)
print(reshaped)
print(reshaped.flatten())
```

## 6ï¸ Mathematical Operations

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
print('Addition:', arr1 + arr2)
print('Subtraction:', arr1 - arr2)
print('Multiplication:', arr1 * arr2)
print('Division:', arr1 / arr2)
```

## 7ï¸ Statistical Functions

```python
arr = np.array([10, 20, 30, 40, 50])
print('Mean:', np.mean(arr))
print('Median:', np.median(arr))
print('Standard Deviation:', np.std(arr))
print('Minimum:', np.min(arr))
print('Maximum:', np.max(arr))
print('Sum:', np.sum(arr))
```

## 8ï¸ Axis Operations

```python
arr = np.array([[1,2,3],[4,5,6]])
print('Sum column-wise:', np.sum(arr, axis=0))
print('Sum row-wise:', np.sum(arr, axis=1))
```

## 9ï¸ Boolean Indexing

```python
arr = np.array([10, 25, 30, 45, 60])
print(arr > 30)
print(arr[arr > 30])
```

## 10 Broadcasting

```python
arr = np.array([1,2,3])
print(arr + 10)

arr2 = np.array([[1],[2],[3]])
arr3 = np.array([10,20,30])
print(arr2 + arr3)
```

## 11 Linear Algebra (Advanced)

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print('Matrix Multiplication:\n', np.dot(A,B))
print('Determinant:', np.linalg.det(A))
print('Inverse:\n', np.linalg.inv(A))
print('Eigenvalues:', np.linalg.eig(A)[0])
```

##  Practice Problems

```python
# Top 3 Highest Marks
marks = np.array([45, 78, 88, 92, 67, 81])
print('Top 3:', np.sort(marks)[-3:])

# Normalize Data
data = np.array([10, 20, 30, 40, 50])
normalized = (data - np.min(data)) / (np.max(data) - np.min(data))
print('Normalized Data:', normalized)

# Even Numbers 1â€“100
numbers = np.arange(1, 101)
even = numbers[numbers % 2 == 0]
print(even)
```
